```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
m=100
X=np.linspace(0,10,m).reshape(m,1)
y=X**2+np.random.randn(m,1)
```

```python
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
model=SVR(C=100)
model.fit(X,y)
model.score(X,y)
plt.plot(X,model.predict(X),c='r')
plt.scatter(X,y)
```
